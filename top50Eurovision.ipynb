```python
from bs4 import BeautifulSoup
```

```python
import requests

r = requests.get('https://eurovision.tv/story/top-50', headers={'user-agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.85 Mobile Safari/537.36'})
r.status_code
```

```python
html = r.content
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
song = [elem.get_text().split('â€“') for elem in soup.find("ol").findAll('p')]
```

```python
song = [s for s in song if len(s) == 2]
    
```

```python
song
```

```python
songs = []
artists = []
for a, s in song:
    print(a, s)
    songs.append(' '.join(s.strip().split()[:-1]))
    artists.append(a.strip())
        
```

```python
artists
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(songs, artists)),
               columns =['Songs', 'Artists'])
df
```
